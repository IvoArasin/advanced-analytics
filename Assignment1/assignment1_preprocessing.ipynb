```python
import pandas as pd
import numpy as np
```

```python
date_columns = ['property_scraped_at', 'host_since', 'reviews_first', 'reviews_last']
my_train = pd.read_csv("train.csv", parse_dates=date_columns)
my_test = pd.read_csv("test.csv", parse_dates=date_columns)
my_train.head()
```

```python
def string_to_int(input_var):
    return_number = 0
    if input_var[0] == 'a':
        return_number = 1
    else:
        return_number = int(input_var.split(" ",1)[0])    
    return return_number
```

```python
from datetime import datetime, timedelta
last_updated_text = my_train.property_last_updated.unique()
#last_updated_text = np.array_str(last_updated_text)
my_list = []
for mytext in last_updated_text: 
    if "week" in mytext:
        xyz = string_to_int(mytext)
        datetime_date = timedelta(weeks=-xyz)
    elif "month" in mytext:
        xyz = string_to_int(mytext) * 4 + 2
        datetime_date = timedelta(weeks=-xyz)
    elif "days" in mytext:
        xyz = string_to_int(mytext)
        datetime_date = timedelta(days=-xyz)
    elif mytext == "yesterday":
        datetime_date = timedelta(days = -1)
    else:
        datetime_date = timedelta()
    my_list.append([mytext, datetime_date])

my_train['property_last_updated'] = my_train['property_last_updated'].apply(lambda x: dict(my_list)[x])
#my_train2['property_scraped_at'] = pd.to_datetime(my_train2['property_scraped_at'])
my_train['property_last_updated_dt'] = my_train['property_scraped_at'] + my_train['property_last_updated']
```

```python
coordinates = my_train[['property_lat', 'property_lon']].values

emb_size = 20
precision = 1e6 

latlon = np.expand_dims(coordinates, axis=-1) 

m = np.exp(np.log(precision) / emb_size) 
angle_freq = m ** np.arange(emb_size) 
angle_freq = angle_freq.reshape(1, 1, emb_size) 

latlon = latlon * angle_freq 
latlon[..., 0::2] = np.cos(latlon[..., 0::2]) 
latlon[..., 1::2] = np.sin(latlon[..., 1::2]) 
latlon = latlon.reshape(-1, 2 * emb_size) 

my_train['exp_latlon1'] = [lat[0] for lat in latlon]
my_train['exp_latlon2'] = [lat[1] for lat in latlon]
```

```python
from haversine import haversine
from sklearn.cluster import KMeans
clustering = KMeans(n_clusters=20, 
                    max_iter=1111, 
                    random_state=228).fit(coordinates)

cluster_centers = {i: tuple(centroid) for i, centroid in enumerate(clustering.cluster_centers_)}
def cluster_features(df):
    for i, cc in enumerate(cluster_centers.values()):
        my_train[f'cluster_{i}'] = my_train.apply(lambda x: haversine((x['property_lat'], x['property_lon']), cc, unit='ft'), axis=1)
    return df
df = cluster_features(my_train)
```
